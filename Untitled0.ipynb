```python
import matplotlib.pyplot as plt
import numpy as np
tf.random.set_seed(25)

y=np.array([16.68, 11.5, 12.03, 14.88, 13.75, 18.11, 8, 17.83, 79.24, 21.5, 40.33, 21, 13.5, 19.75, 24, 29, 15.35, 19, 9.5, 35.1, 17.9, 52.32, 18.75, 19.83])
x=np.array([7, 3, 3, 4, 6, 7, 2, 7, 30, 5, 16, 10, 4, 6, 9, 10, 6, 7, 3, 17, 10, 26, 9, 8])

# Corregir el nombre de la función scatter
plt.scatter(x, y, marker='.')

```

```python

import tensorflow as tf

capa = tf.keras.layers.Dense(units=1, input_shape=[1])
capa2 = tf.keras.layers.Dense(units=2)
capa3 = tf.keras.layers.Dense(units=1, activation='tanh')
modelo = tf.keras.Sequential([capa, capa2, capa3])

```

```python
modelo.compile(
    optimizer = tf.keras.optimizers.Adam(0.1),
    loss = 'mean_squared_error'
)
```

```python
print("iniciando entrenamiento....")
historial = modelo.fit(x,y, epochs=1000,verbose=False)
print("fin")
```

```python
plt.xlabel('Nro epoca')
plt.ylabel('magnitud de periodo')
plt.plot(historial.history['loss'])
```
